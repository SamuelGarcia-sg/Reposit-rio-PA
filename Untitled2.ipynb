1. Número ímpar ou par

```python
numero = int(input("Digite um número: "))

if numero % 2 == 0:
    print(f"O número {numero} é par.")
else:
    print(f"O número {numero} é ímpar.")
```

2. Maior de dois números

```python
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

if num1 > num2:
    print(f"O maior número é: {num1}")
elif num2 > num1:
    print(f"O maior número é: {num2}")
else:
    print("Os números são iguais.")
```

3. aprovação do aluno

```python
nota = float(input("Digite a nota do aluno: "))

if nota >= 7:
    print("Aprovado")
elif nota >= 5 and nota < 7:
    print("Recuperação")
else:
    print("Reprovado")
```

4. Classificação de idade

```python
idade = int(input("Digite a idade: "))

if idade >= 0 and idade <= 12:
    print("Criança")
elif idade >= 13 and idade <= 17:
    print("Adolescente")
elif idade >= 18 and idade <= 59:
    print("Adulto")
elif idade >= 60:
    print("Idoso")
else:
    print("Idade inválida.")
```

5. Senha correta

```python
SENHA_FIXA = "Colab123"

senha_digitada = input("Digite a senha: ")

if senha_digitada == SENHA_FIXA:
    print("Senha correta!")
else:
    print("Senha incorreta.")
```

6. Positivo, negativo ou zero

```python
numero = float(input("Digite um número: "))

if numero > 0:
    print("O número é positivo.")
elif numero < 0:
    print("O número é negativo.")
else:
    print("O número é zero.")
```

7. Calculadora simples

```python
# Escrever um programa que soma dois números
# Entrada
a = int(input("Digite o primeiro número: "))
b = int(input("Digite o segundo número: "))
# Processamento
soma = a + b
subtracao = a - b
divisao = a / b
div_int = a // b
multiplicacao = a * b
resto = a % b
# Saída

print ("|--------------------------------------|")
print ("|--------CALCULADORA SIMPLES-----------|")
print ("|                                      |")
print ("|---------------SOMA-------------------|")
print (f"O resultado da soma de {a} mais {b} é igual a {soma}")
print ("|                                      |")
print ("|-------------SUBTRAÇÃO----------------|")
print (f"O resultado da subtração de {a} menos {b} é igual a {subtracao}")
print ("|                                      |")
print ("|--------------DIVISÃO-----------------|")
print (f"O resultado da divisão de {a} dividido por {b} é igual a {divisao}")
print ("")
print ("|----------DIVISÃO INTEIRA-------------|")
print (f"O resultado inteiro da divisão de {a} dividido por {b} é igual a {div_int}")
print ("|                                      |")
print ("|-----------MULTIPLICAÇÃO--------------|")
print (f"O resultado da multiplicação de {a} vezes {b} é igual a {multiplicacao}")
print ("|                                      |")
print ("|--------------RESTO-------------------|")
print (f"O resto de {a} dividido por {b} é igual a {resto}")
```

8. Dia da semana

```python
numero_dia = int(input("Digite um número de 1 a 7: "))

dias_da_semana = {
    1: "Domingo",
    2: "Segunda-feira",
    3: "Terça-feira",
    4: "Quarta-feira",
    5: "Quinta-feira",
    6: "Sexta-feira",
    7: "Sábado"
}

dia = dias_da_semana.get(numero_dia, "Número inválido. Digite um número entre 1 e 7.")
print(f"O dia correspondente é: {dia}")
```

9. Ano bissexto

```python
ano = int(input("Digite um ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é bissexto.")
else:
    print(f"O ano {ano} não é bissexto.")
```

10. Desconto em compras

```python
valor_compra = float(input("Digite o valor da compra: "))

desconto = 0
if valor_compra >= 100:
    desconto = 0.10
elif valor_compra >= 50 and valor_compra < 100:
    desconto = 0.05

valor_final = valor_compra * (1 - desconto)

print(f"Valor da compra: R${valor_compra:.2f}")
print(f"Desconto aplicado: {desconto*100:.0f}%")
print(f"Valor final com desconto: R${valor_final:.2f}")
```
