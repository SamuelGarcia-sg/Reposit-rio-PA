 QUESTÃO 02 - Soma dos números

```python
# Solicite ao usuário um número inteiro positivo.
# Utilizando laços de repetição, calcule e exiba a soma de todos os números de 1 até o número informado

# Entrada
número = int(input("Digite um número inteiro: "))
soma = 0

for i in range(1, número + 1):
    soma += i

print(f"A soma de 1 até {número} = {soma}")
```

QUESTÃO 03 - Tabuada

```python
# Solicite ao usuário seu número inteiro e exiba a tabuada de 1 até 10

# Entrada
número = int(input("Digite um número inteiro: "))

print(f"Tabuada do {número}:")
for i in range(1, 11):
    resultado = número * i
    print(f"{número} x {i} = {resultado}")
```

QUESTÃO 04 - Fatorial

```python
# Solicite ao usuário um número inteiro positivo e calcule seu fatorial utilizando um laço de repetição

# Entrada
número = int(input("Digite um número inteiro: "))

if número < 0:
    print("Por favor, digite apenas números positivos.")
else:
    fatorial = 1
    for i in range(1, número + 1):
        fatorial *= i

    print(f"O fatorial de {número}! é: {fatorial}.")
```

QUESTÃO 05 - Caixa eletrônico

```python
# Um caixa eletrônico possui notas de: R$ 100, R$ 50, R$ 20, R$ 10, R$ 5, R$ 2, R$ 1
# Solicite ao usuário o valor do saque e informe a menor quantidade possível de notas para realizar o pagamento

# Entrada
saque = float(input("Digite o valor do saque: "))
valor_restante = saque

notas = [100, 50, 20, 10, 5, 2, 1]

print("\n--- Comprovante do Saque ---")
print(f"Valor solicitado: R$ {saque:.2f}")

for nota in notas:
    quantidade = int(valor_restante // nota)

    if quantidade > 0:
        print(f"{quantidade} nota(s) de R$ {nota:.2f}")
        valor_restante %= nota

if valor_restante > 0:
    print(f"Troco não exato: R$ {valor_restante:.2f} restante")
```
