2- contagem regressiva, o usuário digita um número e o programa imprimi do número ate o zero

```python
numero = int(input("Digite um número: "))
soma = 0

for i in range(numero, -1, -1):
    soma += i
    print(i)
```

5- soma, o usuário digita o número e o programa soma os números do zero ate o número


```python
numero = int(input("Digite um número inteiro positivo: "))

if numero < 0:
    print("Por favor, digite um número maior ou igual a zero.")
else:
    soma = (numero * (numero + 1)) // 2

    print(f"A soma de todos os números de 0 até {numero} é: {soma}")
```

6- tabuada, o usuário digita um número e o programa imprimi a tabuada do número

```python
numero = int(input("Digite um número para ver a sua tabuada: "))

print(f"\n--- Tabuada do {numero} ---")

for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```

7- contar letras, o usúario digita uma palavra e o programa conta as letras da palavra

```python
texto = input("Digite uma palavra (ou frase) para contar as letras: ")

texto_sem_espacos = texto.replace(" ", "")

total_letras = len(texto_sem_espacos)

print(f"A palavra/frase digitada tem {total_letras} letras.")
```

8- contar vogais, o usúario digita uma palavra e o programa conta as vogais

```python
texto = input("Digite uma palavra ou frase: ")

vogais = "aáãâeéêiíoóõôuú"

total_vogais = 0

for letra in texto.lower():
    if letra in vogais:
        total_vogais += 1

print(f"O texto digitado contém {total_vogais} vogais.")
```

9- maior número digitado, o usúario digita quatro números e o programa imprimi o maior número digitado

```python
numeros = []

print("Digite quatro números, um de cada vez:")

for i in range(4):
    num = float(input(f"Digite o {i+1}º número: "))
    numeros.append(num)

maior_numero = max(numeros)

print(f"\nO maior número digitado foi: {maior_numero}")
```

10- o usúario digita um número entre 01 e 100 e o programa mostra quais são divisiveis por 3 e 5

```python
numero = int(input("Digite um número entre 1 e 100: "))

if numero < 1 or numero > 100:
    print("Número inválido! Por favor, digite um número que esteja entre 1 e 100.")
else:
    print(f"\nNúmeros de 1 até {numero} que são divisíveis por 3 e 5 ao mesmo tempo:")

    encontrou = False

    for i in range(1, numero + 1):

        if i % 3 == 0 and i % 5 == 0:
            print(i)
            encontrou = True
```
